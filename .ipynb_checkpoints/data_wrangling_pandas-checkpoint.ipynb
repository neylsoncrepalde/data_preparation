## Selecting Columns

To select columns with pandas, you should use a list `[]` inside a subset bracket. Like this:

```python
iris[["SepalLength", "PetalLength"]].head()
```

To exclude some columns, you can use the `drop` method.

```python
iris.drop(columns=["PetalWidth", "SepalWidth"]).head()
```

## Filtering

To filter cases on a pandas DataFrame based in any condition, we can use the `loc` method which is simply a method to do subseting specifying rows and columns. 

The basic syntax is 

`df.loc[rows,columns]`

Attention that `loc` uses brackets, not parenthesis. One can pass a condition to the rows part. Use `:` to get all rows or columns.

```python
iris.loc[:,"Name"].head()
```

```python
# Get only flowers with SepalLenght bigger than or equal 7
iris.loc[iris["SepalLength"] >= 7, :]
```

```python
# Now, the same filter but returning only SepalLength and Name
iris.loc[iris["SepalLength"] >= 7, ["SepalLength", "Name"]]
```

Notice that we pass a condition for row filtering before the comma and a condition for column selection after the comma within a list.

Now, let's do some math to filter. Let's get only the flowers bigger than the mean SepalLength.

```python
m = iris["SepalLength"].mean()
iris.loc[iris["SepalLength"] > m, ["SepalLength", "Name"]]
```

```python
# Or
iris.loc[iris["SepalLength"] >= iris["SepalLength"].mean(), 
         ["SepalLength", "Name"]]
```

# Sorting values

Let's put some order to this data! Let's get only the flowers which have a SepalLength bigger than 7, get SepalLength, PetalWidth and Name and order by PetalWidth.

```python
# \ means line break
iris.loc[iris["SepalLength"] >= 7, 
         ["SepalLength", "PetalWidth", "Name"]].\
sort_values(by="PetalWidth")
```

```python
# The same, now descending
iris.loc[iris["SepalLength"] >= 7, 
         ["SepalLength", "PetalWidth", "Name"]].\
sort_values(by="PetalWidth", ascending=False)
```

# Creating new variables

Let's create a variable according to the SepalLength. If it's bigger than the mean, we will write "Big". If it's equal or less than the mean, we will write "Small". Let's check the distribution for this new variable too.

```python
iris["Size"] = "Small"
iris.loc[iris["SepalLength"] > iris["SepalLength"].mean(), 
        "Size"] = "Big"
print(iris.Size.value_counts())
```

```python
# Vamos plotar O tamanho da pétala cruzando com o tamanho da 
# sépala e colorindo pelo tamanho
fig, ax = plt.subplots()
for s in ["Small", "Big"]:
    if s == "Small":
        ax.scatter(x=iris.loc[iris.Size == s,"SepalLength"], 
                   y=iris.loc[iris.Size == s,"PetalLength"],
                   color = "blue", label = s)
    else:
        ax.scatter(x=iris.loc[iris.Size == s,"SepalLength"],
                   y=iris.loc[iris.Size == s,"PetalLength"],
                   color = "red", label = s)
ax.legend()
plt.title("Scatterplot")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

Now, let's make a 3 categories' variable.

```python
iris.SepalLength.describe()
```

```python
iris["Size2"] = "Small"
iris.loc[(iris["SepalLength"] >= 5.5) & (iris["SepalLength"] < 6.8), 
        "Size2"] = "Medium"
iris.loc[iris["SepalLength"] >= 6.8, 
        "Size2"] = "Big"
print(iris.Size2.value_counts())
```

```python
 # Vamos plotar O tamanho da pétala cruzando com o tamanho da 
# sépala e colorindo pelo tamanho
fig, ax = plt.subplots()
for s in ["Small", "Medium", "Big"]:
    if s == "Small":
        ax.scatter(x=iris.loc[iris.Size2 == s,"SepalLength"], 
                   y=iris.loc[iris.Size2 == s,"PetalLength"],
                   color = "green", label = s)
    elif s == "Medium":
        ax.scatter(x=iris.loc[iris.Size2 == s,"SepalLength"], 
                   y=iris.loc[iris.Size2 == s,"PetalLength"],
                   color = "orange", label = s)
    else:
        ax.scatter(x=iris.loc[iris.Size2 == s,"SepalLength"],
                   y=iris.loc[iris.Size2 == s,"PetalLength"],
                   color = "yellow", label = s)
ax.legend()
plt.title("Scatterplot")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```
